#### Since our purpose of this analysis is to find out which team among inter milan and Man. city is going to perform best on june 10th.

#### So,we're going to exclude other teams from our analysis and mainly focus with man. city and inter milan team.

### Team with great co-ordination or cooperation among its members

```python
attack_df =df_attacking[(df_attacking["club"]=="Man. City") | (df_attacking["club"]=="Inter")]
```

```python
attack_df
```

```python
plt.figure(figsize=(10, 5))
plt.title('Number of assist by teams', fontsize=16, fontweight='bold')

ax=sns.barplot(x=["Inter","Man. City"], y=np.array(attack_df.groupby("club")["assists"].sum()))
# Calculate the average
average_man = attack_df[attack_df["club"]=="Man. City"]["assists"].mean()
average_inter = attack_df[attack_df["club"]=="Inter"]["assists"].mean()
# Add the average line
ax.axhline(average_man, color='orange', linestyle='--', label='Average_man')
ax.axhline(average_inter, color='blue', linestyle='--', label='Average_inter')
plt.ylabel("Total Assists")
plt.yticks(range(1,30,2))
plt.xlabel("Teams")
plt.legend()
plt.grid()
plt.show()
```

#### Team Man. city shows a great coordination or coorporation among their team members as compare to the team members of Inter

### Team with fouls

```python
disciplinary_df =df_disciplinary[(df_disciplinary["club"]=="Man. City") | (df_disciplinary["club"]=="Inter")]
```

```python
disciplinary_df.reset_index(drop=True, inplace=True)
```

```python
disciplinary_df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Team with maximum foul', fontsize=16, fontweight='bold')

ax=sns.barplot(x=["Inter","Man. City"], y=np.array(disciplinary_df.groupby("club")["fouls_committed"].mean()))

plt.ylabel("Average Foul")
plt.yticks(range(1,10))
plt.xlabel("Teams")
plt.grid()
plt.show()
```

#### Man. city is carrying on an average little bit higher number of fouls with them from past seasons,even though their team coordination is high but they suffered more often from  the fouls.It could be possible that their play style is aggresive or they are passionated to win every much they play or they show tons of sportsman energy that's why their coordination is much better then the team Inter.

### Team With high passing accuracy

```python
distributon_df=df_distributon [(df_distributon ["club"]=="Man. City") | (df_distributon["club"]=="Inter")]
```

```python
distributon_df.reset_index(drop=True,inplace=True)
```

```python
distributon_df= distributon_df.sort_values(by='pass_accuracy', ascending=False)
```

```python
distributon_df.head()
```

#### From above dataset,Egan-Riley who is from team Man. City has shown an outstanding passing accuracy of 95%.

```python
plt.figure(figsize=(10, 7))
plt.title('Team with high Passing Accuracy', fontsize=16, fontweight='bold')

perc=np.array(distributon_df.groupby("club")["pass_completed"].mean())

plt.pie(x=perc,labels=["Inter","Man. City"],autopct='%1.1f%%')

plt.show()
```

#### Since we know that members of Team Man. City has very good coordination among each other,which ultimately help them to perform better when it comes to passing football or receiving it from other members.

### Team with Great Defender

```python
defending_df=df_defending[(df_defending["club"]=="Man. City") | (df_defending["club"]=="Inter")]
```

```python
defending_df.reset_index(drop=True,inplace=True)
```

```python
defending_df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.title("Team's Defending Performance", fontsize=16, fontweight='bold')

x_a = ["Inter", "Man. City"]
width = 0.35  # Width of the bars

# Calculate the position of the bars
x_pos = np.arange(len(x_a)) #[0,1]

# Get the mean values for balls_recovered and tackles for each club
balls_recovered_means = defending_df.groupby("club")["balls_recoverd"].mean()
tackles_means = defending_df.groupby("club")["tackles"].mean()

# Create the bar plots
ax = plt.bar(x_pos, balls_recovered_means, width, label="Balls Recovered")
ax2 = plt.bar(x_pos + width, tackles_means, width, label="Tackles")

# Set x-axis ticks and labels
plt.xticks(x_pos + width/2, x_a)

plt.ylabel("Total")
plt.xlabel("Teams")
plt.legend()  # Show legend

plt.show()
```

#### On average, the players from Man. City exhibit a higher ball recovery rate, while Inter demonstrates superior tackling skills. This discrepancy in tackling ability potentially contributes to Inter committing fewer fouls compared to Man. City. Conversely, Man. City has a player or players with slightly lower average tackling skills, which could increase the likelihood of fouls being committed. As observed in our previous analysis, Man. City tends to have a higher frequency of fouls. However, it's worth noting that the average tackling skills between the two teams are quite similar. Consequently, this factor may have minimal impact, and overall, Man. City has displayed a stronger performance.
